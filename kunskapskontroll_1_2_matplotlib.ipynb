### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
Figure:  This is the whole figure and can be seen as a canvas.
Axes:    The part of the 'canvas' that the plot is attached on.We call methods that do the plotting directlyfrom the Axes which gives high flexibility in customizing our plots.
Axis:    Sets the scale and limits and generate trics (the marks on the Axis) and trickables (strings labeling the tricks).
Artist:  Artist is everything which is visible on the figure like figure, axes and axis. Most Artist are tied to an Axes.  
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python

# Generate x values using np.linspace
x = np.linspace(-4, 4, 100)  # 100 points from -4 to 4

# Calculate y values
y = x**2

# Object-Oriented Approach
fig, ax = plt.subplots()
ax.plot(x, y, label='y = x^2')
ax.set_title('Plot of y = x^2')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()

# pyplot Approach
plt.figure()
plt.plot(x, y, label='y = x^2')
plt.title('Plot of y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Create a figure with two subplots
plt.figure(figsize=(12, 5))

# First subplot: Scatter plot
plt.subplot(1, 2, 1)
plt.scatter(random_data_x, random_data_y, alpha=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')

# Second subplot: Bar plot
plt.subplot(1, 2, 2)
plt.bar(names, values)
plt.xlabel('Fruits')
plt.ylabel('Quantity')
plt.title('Bar Plot')

plt.tight_layout()  # To ensure the subplots don't overlap

plt.show()
```
